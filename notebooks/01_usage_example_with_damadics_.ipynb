# Experimento

## Carregamento DAMADICS benchmark

```python
def get_fault_data(data, vars_names, ID, size=500):
    fidxs = np.where(data[:, -1] != 0)[0]
    names = vars_names + ['ID']
    
    d = pd.DataFrame(data[fidxs, :-1][:size, :], columns=vars_names)
    d['ID'] = np.repeat(ID, repeats=size)
    
    return d

operations = ['normal', 'f01b']
vars_names, data_all = benchmarks.fetch_damadics(operations).values()

num_points, vlines = 500, []

normal = pd.DataFrame(data_all['normal'][:600, :-1], columns=vars_names[:-1])
normal['ID'] = np.repeat(1, repeats=600)
vlines.append(normal.shape[0] - 1 - 100)

f1 = get_fault_data(data_all['f01b'], vars_names[:-1], 2, size=num_points)
vlines.append(vlines[-1] + f1.shape[0])

vlines.append(vlines[-1] + normal.shape[0])
```

## Aplicação de filtro de média móvel

```python
w = 100

data = pd.concat((normal, f1, normal), ignore_index=True)
data_orig = data.iloc[w:, :].copy().reset_index(drop=True)

for v in ['CV', 'P1', 'P2', 'X', 'F', 'T1']:
    data.loc[w-1:, v] = moving_average(data[v].values, w)
    
data = data.iloc[w:, :].reset_index(drop=True)
```

```python
plt.close('all')

def show_data(data, title=None):
    
    fig, ax = plt.subplots(1, 1, sharex=True, figsize=(6, 4))
    
    if title is not None:
        fig.suptitle(title)
    
    ax = sns.lineplot(data=data[['CV', 'P1', 'P2', 'X', 'F', 'T1']], hue='grupo', legend='full', lw=1, ax=ax)
    labels = [tex.get_text() for tex in ax.get_yticklabels()]
    ax.set_yticklabels(np.round(ax.get_yticks(), 1))
    ax.set_xticks([1] + vlines)
    ax.set_xticklabels([1] + [v + 1 for v in vlines])
    ax.set_ylabel('Amplitude')
    ax.set_xlabel('Tempo $(t)$')
    ax.grid(True, ls=':', alpha=0.8, lw=0.7)

    ax2 = ax.twiny()
    ax2.set_xticks([0] + vlines)
    ax2.set_xticklabels(['Normal', '$f_1$', 'Normal'])
    ax2.set_xlim(ax.get_xlim())

show_data(data_orig, title='Dados Originais')
show_data(data, title='Dados com Filtro')
```

## Execução

```python
colors = cm.rainbow(np.linspace(0, 1, 4))
variables = ['CV', 'P1', 'P2', 'X', 'F', 'T1']

def show_clus_contrib_result(data, method_name, color):
  
    grid_dict = {'alpha': 0.4, 'ls': ':', 'lw': .7, 'color': 'k'}

    fig, ax1 = plt.subplots(figsize=(8, 2))        

    # Method classification.
    ax1.plot(data.ID, ls='-', label='Real', lw=1.5, color='g')
    ax1.plot(data[method_name + '_out'], ls='--', label=method_name, lw=1.5, color=color)
       
    max_id = data[method_name + '_out'].max()
    
    ax1.set_xticks([0] + vlines)
    ax1.set_xticklabels([1] + [v + 1 for v in vlines])
    ax1.grid(**grid_dict)
    ax1.set_xlabel('Tempo $(t)$')
    ax1.set_ylabel('Grupo ID')
    yticks = np.linspace(1, max_id, 10, dtype=int)
    ax1.set_yticks(yticks)
    ax1.set_ylim(0.9, max_id + 0.1)
    ax1.grid(**grid_dict)
    ax1.legend(fontsize=8)
    
    # Contributions.
    plt.savefig(f'res_damadics_{method_name}_.pdf', format='pdf')

    fig2, ax2 = plt.subplots(figsize=(8, 4))        

    sns.heatmap(data[diag_names][data[diag_names].sum(axis=1) > 0].T, cmap='Blues', 
                linewidths=0.01, linecolor='white', yticklabels=variables, 
                cbar_kws={'shrink': 0.8, 'fraction':0.1, 'label': 'Contribuição (%)'}, ax=ax2)
    
    ax2.set_xlabel('Instante de criação/transição de grupo $(t)$')
    ax2.set_ylabel('Variável')
    ax2.grid(**grid_dict)
    plt.show()
    
```

### eSBM+

```python
name_method = 'eSBM+'
out_name = name_method + '_out'
diag_names = [name_method + f'_diag_{v}'  for v in variables]

method_names = [name_method]

param_evolving = {
    'dim': 6,
    'k': 15,
    'gamma_group': 0.99,
    'gamma_point': 0.05,
    'theta': 0.06,
    'tau': 1e-5,
    'sml_func_name': 'lk'
}

param_rbc = {
    'w_train': 50,
    'w_diag': 1,
    'n_pc' : 0.95,
    'conf_level': 0.99,
    'index': 'combined',
    'lags': 0
}

evolving_m = clustering.esbm.eSBMPlus
rbc = contribution.RBC

monitor = Monitor(clustering_method=(evolving_m, param_evolving), diagnosis_method=(rbc, param_rbc))

data[out_name] = np.zeros(shape=data.shape[0], dtype=np.uint8)

for d_name in diag_names:
    data[d_name] = np.zeros(shape=data.shape[0], dtype=np.float64)

for i, row in data.iterrows():
    x = row[variables].values[None, :].astype(np.float64)
    
    status, cstar, contrib = monitor.process_input(x)
    data.loc[i, out_name] = cstar
    
    if contrib is not None:
        data.loc[i, diag_names] = contrib[0]
```

```python
show_clus_contrib_result(data, name_method, colors[0])
```
